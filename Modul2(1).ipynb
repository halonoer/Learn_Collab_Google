```python
def survival_stacked_bar(variable):
    died = data[data['Survived'] == 0][variable].value_counts() / len(data[data['Survived'] == 0])
    survived = data[data["Survived"] == 1][variable].value_counts() / len(data[data['Survived'] == 1])

    dataset = pd.DataFrame([died, survived])
    dataset.index = ['Meninggal', 'Selamat']
    dataset.plot(kind="bar", stacked=True, title='Prosentase')

    return dataset.head()
```

```python
survival_stacked_bar('Sex')
```

```python
def survival_stacked_bar_age():
    # Mengelompokkan usia ke dalam kategori
    bins = [0, 18, 60, 100]  # Rentang usia
    labels = ['Anak', 'Dewasa', 'Lansia']  # Label kelompok usia

    # Tambahkan kolom 'AgeGroup' berdasarkan rentang usia
    data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels)

    # Menghitung jumlah yang meninggal dan selamat berdasarkan kelompok usia
    died = data[data['Survived'] == 0]['AgeGroup'].value_counts(normalize=True)
    survived = data[data["Survived"] == 1]['AgeGroup'].value_counts(normalize=True)

    # Membuat DataFrame dengan data tersebut
    dataset = pd.DataFrame([died, survived])
    dataset.index = ['Meninggal', 'Selamat']

    # Plot diagram batang bertumpuk
    dataset.plot(kind="bar", stacked=True, title='Prosentase Penumpang Berdasarkan Kelompok Usia')

    return dataset.head()

# Memanggil fungsi untuk menampilkan data berdasarkan kelompok usia
survival_stacked_bar_age()
```
